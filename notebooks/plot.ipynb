```python
import matplotlib.pyplot as plt
import json



```

```python
with open("results.json") as f:
    data = json.load(f)

plant_dates = [i for i in data["height"].keys()]
plant1_height = [i[5] for i in data["height"].values()]
plant1_area = [i[5] for i in data["area"].values()]
```

```python
plt.plot(plant_dates, plant1_area)
plt.title("Plant Height Over Time")
plt.xlabel("Date")
plt.ylabel("Height (mm)")
plt.show()
```

```python
plt.plot(plant_dates, plant1_height, "-o")
plt.title("Plant Height Over Time")
plt.xlabel("Date")
plt.ylabel("Height (mm)")
plt.show()
```
