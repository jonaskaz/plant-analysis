```python
import os
from dotenv import load_dotenv
from image_helper import ImageHelper
from datetime import datetime

load_dotenv()

base_url = os.environ.get("STORAGE_URL")
# base_url will be empty if .env file is missing
assert base_url != ""
im_helper = ImageHelper(base_url)
dt_str = '3/12/2022 19:08:18'
dt = im_helper.dt_from_string(dt_str)
assert(im_helper.get(dt, "top"))
```

```python
import shutil
from time import sleep
image_dir = "side_images/"
image_dates = ["30/11", "1/12", "2/12", "3/12", "4/12", "5/12"]
for date in image_dates:
    for time in range(1, 24):
        dt_str = f'{date}/2022 {time}:00:00'
        dt = im_helper.dt_from_string(dt_str)
        if im_helper.get(dt, "side"):
            save_path = image_dir + dt_str.replace("/", "-") + ".png"
            print(save_path)
            shutil.copyfile(im_helper.image.name, save_path)
        else:
            print(f"Could not get image for: {dt_str}")

            
```
